# VD17-Bereiningung Druckjahr Teil I

Bei Datensaetzen, die aus historischen Quellen entstehen, sind Angaben wie Datum, Jahr, Ort usw. selten einheitlich. Um sie auszuwerten muessen die Spalten bereinigt, standardisiert bzw. normalisiert werden.  In diesem Jupyter-Notebook lernen wir, wie die Druckjahr-Spalte im VD17 bereinigt werden kann. 

1. Wir Importieren 'Pandas', die Standard-Bibliothek fuer Datenanalyse in Python:

```python
import pandas as pd
```

2. Im zweiten Schritt laden wir die VD17-Daten in einer Variablen, hier 'vd17', und zwar als Dataframe. Ein Dataframe ist eine Art Tabelle, die voruebergehend aus den Daten entsteht, wenn wir pandas, hier als 'pd' abgekuerzt, anwenden.  

vd17: Variable in der die Daten gespeichert sind
pd.read_csv: Pandas-Funktion: Hier sagen wir zum Computer, welche Datei er als DataFrame strukturieren soll 
('VD17_Sample_April23.csv'): Die Datei mit der wir arbeiten wollen. In Fall dieses Notebooks geben wir den Namen der Datei, die bei Binder gespeichert ist. Im lokalen Computer nutzen wir den Computer-Pfad, der normalerweise mit "C:\Users\Name" anfaengt

```python
vd17 = pd.read_csv('VD17_Sample_April23.csv')
```

3. Damit koennen wir jezt schauen, wie die die Druckjahr-Spalte aussieht:

```python
vd17['Druckjahr']
```

4. Wie wir gesehen haben, enthaelt die Druckjahr-Spalten viele 'nicht nummeriche Zeichen', wie "i.e.", "erschienen", eckige Klammer usw. Diese Angaben koennen wir nicht auswerten, also diese Zeilen muessen bereinigt werden. Um sicher zu stellen, dass die historischen Daten nicht verloren gehen,  erstellen wir eine neue Spalte:

```python
vd17['DruckjahrBerein'] = vd17.loc[:, 'Druckjahr']
```

4. Hier ueberpruefen wir, ob es geklappt hat:

```python
vd17['DruckjahrBerein']
```

5. Mit 'vd17.head(20)' werden die ersten 20 Zeilen des vd17 Dataframes geladen. Damit kann man sehen, dass die Spalte 'DruckjahrBerein' am Ende der Tabelle entstanden ist. 

```python
vd17.head(20)
```

6. Mit der Spalte "DruckBerein" arbeiten wir jetzt. Mit der Funktion str.replace und "regular expressions, regex" entfernen zunaechst alle "nicht-numerischen" Zeichen:

```python
vd17['DruckjahrBerein'] = vd17['DruckjahrBerein'].str.replace(r'\D', '', regex = True)
```

Hat es geklappt? 

```python
vd17['DruckjahrBerein']
```

7. wenn es geklappt hat, entsteht in Zeile 3 und 4 eine 8-stellige Nummer. Sie besteht aus zwei Jahreszahlen. Fuer uns interessant ist nur das erste Jahr, d.h. wir muessen diese Nummer spalten. Das geht bei Pandas so:

vd17['DruckjahrBerein']: Di Spalte 'DruckjahrBerein' in vd17 
.str.slice(stop = 4): wird so aufgeteilt, dass nur die ersten 4 Zeichen in der Spalten gespeichert werden.

Mit "vd17['DruckjahrBerein']" ueberpruefen wir, ob es funktioniert hat. 

```python
vd17['DruckjahrBerein'] = vd17['DruckjahrBerein'].str.slice(stop = 4)
vd17['DruckjahrBerein']
```

8. In der Spalte "DruckjahrBerein" haben wir jetzt eine 4-stellige Nummer, die unser Druckjar ist. Damit ist die Spalte fertig fuer die Ueberpuefung und Weiterbereinigung in Excel. Wir speichern die Tabelle in Binder und laden sie herunter. 

```python
vd17.to_csv(r'VD17_DHGruppe.csv', encoding='utf-8')
```
